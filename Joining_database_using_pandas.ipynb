```python
# when do you used left_on, right_on when you merge? when you have different names for the columns
```

```python
# let us change the movieId name
movies.rename(columns={"movieId":"m_id"},inplace=True)
```

```python
movies.columns
```

```python
# if we try to merge with different titles, it will return error
pd.merge(movies, ratings).head()
```

```python
# now let us merge
pd.merge(movies, ratings, left_on='m_id', right_on='movieId').head()
```

```python
# let us join on one index
# let us set the index for movies on "m_id"
movies=movies.set_index('m_id')
movies.head()
```

```python
pd.merge(movies, ratings, left_index=True, right_on="movieId").head()
```

```python
# what if we want to merge on two indexes? 
ratings= ratings.set_index('movieId')
ratings.head()
```

```python
pd.merge(movies, ratings, left_index=True, right_index=True).head()
```

Four Types of Joints

```python
A= pd.DataFrame({'color': ['green','red','yellow'],
                 'num':[1,2,3]})
A
```

```python
B = pd.DataFrame({'color':['green','pink','yellow'],
                  'size':['S','M','L']})
B
```

### Inner
#### found in both

```python
pd.merge(A, B,how='inner')
```

### Outer
#### Found in either one of them

```python
pd.merge(A, B, how='outer')
```

### Left
#### Include all observations from left

```python
pd.merge(A, B, how='left')
```

### Right 
#### Include all observations from right

```python
pd.merge(A, B, how='right')
```

### Merge_asof

```python
>>> left = pd.DataFrame({'a': [1, 5, 10], 'left_val': ['a', 'b', 'c']})
>>> left
```

```python
>>> right = pd.DataFrame({'a': [1, 2, 3, 6, 7],
...                       'right_val': [1, 2, 3, 6, 7]})
>>> right
```

```python
pd.merge(left,right,on='a')
```

```python
pd.merge_asof(left,right,on='a')
```
